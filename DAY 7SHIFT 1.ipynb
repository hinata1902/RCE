```python
from abc import ABC,abstractclassmethod




class Demo(ABC):
    @abstractclassmethod
    def show(self):
        pass


class Child(Demo):
    def display(self):
        print("this is child method")
    def show(self):
        print("this is parent method")

c=Child()
c.show()
c.display()    



```

```python
#question 2

class Rectangle:
    def __init__(self,length,breadth):
        self.length=length
        self.breadth=breadth

    def area(self):
        return("area=",self.length*self.breadth)

    def parimeter(self):
        return("parimeter=",2*(self.length+self.breadth))

class Square(Rectangle):
    def __init__(self,s):
        self.s=s
        super().__init__(s,s)
    def areas(self,s):
        return("area of square=",self.s*self.s)
        

    
    def parimeters(self,s):
        return("area of square=",4*self.s)

        
 

s=int(input("enter amount if withdraw else enter 0"))
ob=Square(s)
ob.area()
ob.parimeter()
ob.areas()
ob.parimeters()

```
